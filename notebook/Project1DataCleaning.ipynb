```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

The dataset in converted into two sets based on Region from 122th index , we can make a new column based on the region
1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"
Add new Column with region


```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
df
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values....
df = df.dropna().reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
## remove the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python

## Fix spaces in column
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
## Change the required column as interger data type
df[['month','day','year','Temperature','RH']] = df[['month','day','year','Temperature','RH']].astype(int)
```

```python
df.info()
```

```python
## changeing the other columns to float data datatype
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
  if i!='Classes':
    df[i] = df[i].astype(float)
```

```python
## Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize = (13,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of Sidi-Bel Regions',weight='bold')
```

```python
## Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize = (13,6))
sns.set_style('whitegrid')
sns.countplot(x = 'month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire Analysis of Brajaia Regions',weight='bold')
```
